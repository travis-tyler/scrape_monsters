```python
import pandas as pd
import re
```

```python
df = pd.DataFrame(pd.read_csv('../output/monster_stats_cleaned_1.csv'))
df.head()
```

```python
# delete 'info' where 'actions' duplicate
m = (df['info'].str.contains('Actions |', regex=False))
df.loc[m, 'info'] = ''
```

```python
# break saving throws into columns
df['saving_throws'] = df.saving_throws.fillna('')
df['saving_throws'] = df.saving_throws.str[:-1] + ','

str_save = []
dex_save = []
con_save = []
int_save = []
wis_save = []
cha_save = []

for index, row in df.iterrows():
    if 'Str' in row.saving_throws:
        strength = re.search('Str (.*?)\,', row.saving_throws).group(1)
    else:
        strength = ''
    if 'Dex' in row.saving_throws:
        dex = re.search('Dex (.*?)\,', row.saving_throws).group(1)
    else:
        dex = ''
    if 'Con' in row.saving_throws:
        con = re.search('Con (.*?)\,', row.saving_throws).group(1)
    else:
        con = ''    
    if 'Int' in row.saving_throws:
        intel = re.search('Int (.*?)\,', row.saving_throws).group(1)
    else:
        intel = ''
    if 'Wis' in row.saving_throws:
        wis = re.search('Wis (.*?)\,', row.saving_throws).group(1)
    else:
        wis = ''
    if 'Cha' in row.saving_throws:
        cha = re.search('Cha (.*?)\,', row.saving_throws).group(1)
    else:
        cha = ''
        
    str_save.append(strength)
    dex_save.append(dex)
    con_save.append(con)
    int_save.append(intel)
    wis_save.append(wis)
    cha_save.append(cha)
        
df['str_save'] = str_save
df['dex_save'] = dex_save
df['con_save'] = con_save
df['int_save'] = int_save
df['wis_save'] = wis_save
df['cha_save'] = cha_save

df = df.drop(columns=['saving_throws'])
df.head()
```

```python
# break mon_type into 2 columns
df = df.join(pd.DataFrame(df.mon_type.str.split(' ',1).tolist(),
                                 columns = ['size','monster_type']))
df['monster_type'] = df['monster_type'].str.capitalize()

df = df.drop(columns=['mon_type'])
df.head()
```

```python
# break skill into columns
monster_skills = ['History',
 'Perception',
 'Stealth',
 'Persuasion',
 'Insight',
 'Deception',
 'Arcana',
 'Religion',
 'Acrobatics',
 'Athletics',
 'Intimidation']

df['skills'] = df['skills'].fillna('') + ','

all_skills = []
for index, row in df.iterrows():
    skills_dict = {}
    for skill in monster_skills:
        if skill in row.skills:
            skills_dict[skill] = int(row.skills.split(f'{skill} +')[1].split(',')[0])
        else:
            skills_dict[skill] = 0
    all_skills.append(skills_dict)
    
df = df.join(pd.DataFrame(all_skills))

df.head()
```

```python
# break speed into categories 
speed_types = ['speed','swim','fly','climb','burrow']

df['speed'] = df['speed'].fillna('') + ','

all_speeds = []

for index, row in df.iterrows():
    speed_dict = {}
    for speed in speed_types:
        row_speeds = row.speed.lower()
        if speed in row_speeds:
            if 'speed swim' in row_speeds:
                speed_dict[speed] = 0
            else:
                speed_num = re.search(rf'{speed} (.*?) ft', row_speeds).group(1)
            speed_dict[speed] = int(speed_num)
        else:
            speed_dict[speed] = 0
    all_speeds.append(speed_dict)
    
df = df.drop(columns=['speed'])
df = df.join(pd.DataFrame(all_speeds))

df.head()
```

```python
# break actions into columns 
df['actions'] = df.actions.fillna(' | ')

actions = pd.Series(df.actions.str.split(' \| '))

all_actions = []
for i in actions:
    action_num = len(i)
    empty_num = 10 - action_num
    for x in range(empty_num):
        i.append('')
    all_actions.append(i)

df = df.join(pd.DataFrame(all_actions, columns=['action_1','action_2',
                                                       'action_3','action_4',
                                                       'action_5','action_6',
                                                       'action_7','action_8',
                                                       'action_9','action_10']))

df.head()
```

```python
# break legendary_actions into columns 
df['legendary_actions'] = df.legendary_actions.fillna(' | ')

leg_actions = pd.Series(df.legendary_actions.str.split(' \| '))

all_leg_actions = []
for i in leg_actions:
    leg_action_num = len(i)
    empty_num = 5 - leg_action_num
    for x in range(empty_num):
        i.append('')
    all_leg_actions.append(i)

df = df.join(pd.DataFrame(all_leg_actions, columns=['legend_action_1','legend_action_2',
                                                'legend_action_3','legend_action_4',
                                                'legend_action_5']))

df.head()
```

```python
column_order = ['name', 'size',
       'monster_type', 'alignment', 'ac', 'hp', 'strength', 'str_mod', 'dex',
       'dex_mod', 'con', 'con_mod', 'intel', 'int_mod', 'wis', 'wis_mod',
       'cha', 'cha_mod', 'senses', 'languages', 'cr', 'info',
       'str_save', 'dex_save', 'con_save', 'int_save', 'wis_save', 'cha_save',
       'speed', 'swim', 'fly', 'climb', 'burrow', 'info',
       'number_legendary_actions', 
       'History', 'Perception', 'Stealth', 'Persuasion',
       'Insight', 'Deception', 'Arcana', 'Religion', 'Acrobatics', 'Athletics', 'Intimidation', 
       'action_1', 'action_2', 'action_3', 'action_4', 'action_5', 'action_6', 'action_7',
       'action_8', 'action_9', 'action_10', 'legend_action_1',
       'legend_action_2', 'legend_action_3', 'legend_action_4',
       'legend_action_5']
df = df[column_order]
```

```python
df.to_csv('../output/cleaned_monsters2.csv',index=False)
```
