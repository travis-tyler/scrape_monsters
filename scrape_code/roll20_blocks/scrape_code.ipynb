```python
import selenium
from selenium import webdriver
from selenium.webdriver.common.action_chains import ActionChains
import time
```

```python
url = 'https://roll20.net/compendium/dnd5e/Monsters%20List#content'
```

```python
def browser():
    driver = webdriver.Chrome()
    driver.get(url)
    return driver
```

```python
def main():
    '''open all expansion tabs and add data to dictionary'''
    driver = browser()
    time.sleep(5)
    buttons = driver.find_elements_by_class_name('dropdown-toggle')
    for button in buttons[7:]:
        time.sleep(.5)
        driver.execute_script("arguments[0].click();", button)
        
    monster_cards = driver.find_elements_by_class_name('card')
    
    return monster_cards
```

```python
monster_cards = main()
```

```python
len(monster_cards)
```

```python
moster_stats = []
for card in monster_cards:
    monster = {}
    
    monster_stats = card.text.split('\n')
    name = monster_stats[0]
    mon_type = monster_stats[1].split(', ')[0]
    alignment = monster_stats[1].split(', ')[1]
    ac = monster_stats[2].split(' ')[2]
#     armor_type = monster_stats[2].split('(')[1][:-1]
    hp = monster_stats[3].split(' ')[2]
    speed = monster_stats[4]
    strength = monster_stats[6].split(' ')[0]
    str_mod = monster_stats[6].split('(')[1][:2]
    dex = monster_stats[7].split(' ')[0]
    dex_mod = monster_stats[7].split('(')[1][:2]
    con = monster_stats[8].split(' ')[0]
    con_mod = monster_stats[8].split('(')[1][:2]
    intel = monster_stats[9].split(' ')[0]
    int_mod = monster_stats[9].split('(')[1][:2]
    wis = monster_stats[10].split(' ')[0]
    wis_mod = monster_stats[10].split('(')[1][:2]
    cha = monster_stats[11].split(' ')[0]
    cha_mod = monster_stats[11].split('(')[1][:2]
    other_stats = monster_stats[12:]

    monster['name'] = name
    monster['mon_type'] = mon_type
    monster['alignment'] = alignment
    monster['ac'] = ac
#     monster['armor_type'] = armor_type
    monster['hp'] = hp
    monster['speed'] = speed
    monster['strength'] = strength
    monster['str_mod'] = str_mod
    monster['dex'] = dex
    monster['dex_mod'] = dex_mod
    monster['con'] = con
    monster['con_mod'] = con_mod
    monster['intel'] = intel
    monster['int_mod'] = int_mod
    monster['wis'] = wis
    monster['wis_mod'] = wis_mod
    monster['cha'] = cha
    monster['cha_mod'] = cha_mod
    monster['other_stats'] = ' | '.join(other_stats)
    
    moster_stats.append(monster)

moster_stats
```

```python
import pandas as pd
monster_df = pd.DataFrame(moster_stats)
monster_df
```

```python
monster_df.to_csv('../../output/all_monster_stats')
```
