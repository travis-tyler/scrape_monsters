```python
import pandas as pd
import re
```

```python
df = pd.DataFrame(pd.read_csv('../../output/all_monster_stats'))
df = df.loc[:, ~df.columns.str.contains('^Unnamed')]
df.head()
```

```python
saving_throws=[]
skills=[]
senses=[]
languages=[]
cr=[]
info=[]
actions=[]
num_legendary_actions = []
legendary_actions=[]

for index, row in df.iterrows():
    
    if row.other_stats.startswith('Saving Throws'):
        save_throw = re.search('Saving Throws (.*?)\|\s', row.other_stats).group(1)
    else:
        save_throw = ''
        
    if '| Skills ' in row.other_stats:
        skllz = re.search(' \| Skills (.*?) \|\s', row.other_stats).group(1)
    else:
        skllz = ''
        
    if '| Senses ' in row.other_stats:
        sense = re.search(' \| Senses (.*?) \|\s', row.other_stats).group(1)
    else:
        sense = ''
        
    if '| Languages ' in row.other_stats:
        lang = re.search(' \| Languages (.*?) \|\s', row.other_stats).group(1)
    else:
        lang = ''
        
    if '| Challenge ' in row.other_stats:
        challenge = row.other_stats.split('| Challenge ')[1].split(' (')[0]
    else:
        challenge = ''
        
    if '| Actions |' in row.other_stats:
        if '| Legendary Actions |' in row.other_stats:
            action = re.search('\| Actions \| (.*) \| Legendary Actions \|', 
                               row.other_stats).group(1)
            leg_action = row.other_stats.split('legendary actions at the start of their turn. | ')[1]
            num_legendary = len(leg_action.split('|'))
        else:
            action = row.other_stats.split('| Actions | ')[1]
            leg_action = ''
            num_legendary = 0
    else:
        action = ''
        num_legendary = 0

    # TODO: figure out why regex not working
    other_stats = row.other_stats.split('XP) | ')[1].split(' | Actions |')[0]
#     other_stats = re.search('XP\) \| (.*) \| Actions', row.other_stats).group(1)
         
    saving_throws.append(save_throw)
    skills.append(skllz)
    senses.append(sense)
    languages.append(lang)
    cr.append(challenge)
    info.append(other_stats)
    actions.append(action)
    num_legendary_actions.append(num_legendary)
    legendary_actions.append(leg_action)

df['saving_throws']=saving_throws
df['skills']=skills
df['senses']=senses
df['languages']=languages
df['cr']=cr
df['info']=info
df['actions']=actions
df['number_legendary_actions']=num_legendary_actions
df['legendary_actions']=legendary_actions

df
```

```python
df = df.drop(columns=['other_stats'])
df.columns
```

```python
df.to_csv('../../output/monster_stats_cleaned_1.csv', index=False)
```
